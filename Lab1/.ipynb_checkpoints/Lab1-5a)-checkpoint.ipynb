```python
import qiskit
import numpy as np
from qiskit import *
from qiskit.visualization import plot_histogram, plot_bloch_multivector
from math import sqrt, pi

apitoken = '0ac08bc0e23b22dcff4888822d56241be15b0a64e77e94c42639a524404aafbc7df008ab6043e4706b3b82aa42f03c06464b3c253b92ddf1337e80048cf53be0'
IBMQ.save_account(apitoken)
```

```python
n = 5
q = QuantumRegister(n)
c = ClassicalRegister(n)
circuit = QuantumCircuit(q,c)
```

```python
circuit.h(q)
circuit.measure(q, c)
circuit.draw()
```

```python
simulator = Aer.get_backend('qasm_simulator')
job = execute(circuit, simulator, shots = 8192)
result = job.result()
counts = result.get_counts()
plot_histogram(counts)
```
